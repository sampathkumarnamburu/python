# EDA - Explorarory data Analysis

```python
#list of categorical variable

cat_var = ['Geography','Gender','HasCrCard','IsActiveMember']

#create figure of subplot
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (5,10))
axs = axs.flatten()

#creat barplot

for i, var in enumerate(cat_var):
    sns.countplot(x=var, hue='Exited', data =df, ax = axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)
    
    
#adjusting space
fig.tight_layout()

#show the plot
plt.show()
```

```python
#box plots

num_vars = ['CreditScore', 'Age','Tenure', 'Balance','NumOfProducts','EstimatedSalary']


fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
axs = axs.flatten()

for i,var in enumerate(num_vars):
    sns.boxplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
#label encoding

for col in df.select_dtypes(include=['object']).columns:
    print(f'{col}: {df[col].unique()}')
```

```python
from sklearn import preprocessing
```

```python
for col in df.select_dtypes(include=['object']).columns:
    label_encode = preprocessing.LabelEncoder()
    label_encode.fit(df[col].unique())
    df[col] = label_encode.transform(df[col])
    
    print(f"{col}: {df[col].unique()}")
```

```python
#correlate heatmap
plt.figure(figsize=(20,16))
sns.heatmap(df.corr(), cmap='Greens', annot=True)
```
